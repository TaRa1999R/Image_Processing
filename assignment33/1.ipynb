```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_flower = cv2.imread ("inputs\input_1_BW_1.jpg" , cv2.IMREAD_GRAYSCALE)
img_f_hist = cv2.calcHist ([img_flower] , [0] , None , [256] , [0 , 256])

plt.axis ('off')
plt.title ("Flower Photo")
plt.imshow (img_flower , cmap='gray')
plt.show ()
```
