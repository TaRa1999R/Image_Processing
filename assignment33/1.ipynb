```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_flower = cv2.imread ("inputs\input_BW.jpg" , cv2.IMREAD_GRAYSCALE)
img_f_hist = cv2.calcHist ([img_flower] , [0] , None , [256] , [0 , 256])

img =  cv2.imread ("inputs\input_BW_3.jpg" , cv2.IMREAD_GRAYSCALE)
img_hist = cv2.calcHist ([img_flower] , [0] , None , [256] , [0 , 256])

plt.subplot (2 , 2 , 1)
plt.title ("Flower Photo")
plt.imshow (img_flower , cmap='gray')

plt.subplot (2 , 2 , 2)
plt.plot (img_f_hist)
plt.title ("Flower Histogram")

plt.subplot (2 , 2 , 3)
plt.title ("B&W Photo")
plt.imshow (img , cmap='gray')

plt.subplot (2 , 2 , 4)
plt.plot (img_hist)
plt.title ("B&W Histogram")
plt.show ()


```
